## 5. Checking duplicated data

Trong quá trình kiểm tra tính nhất quán của dữ liệu, ta sử dụng phương thức `duplicated()` kết hợp với `any()` của Pandas để phát hiện các dòng trùng lặp. Biến lưu kết quả kiểm tra (`duplicated_daily` và `duplicated_summary`) có giá trị **True** nếu tồn tại các dòng trùng lặp, và **False** nếu dữ liệu hoàn toàn duy nhất.

```python
duplicated_daily = df_daily.duplicated(subset = ["country", "date"]).any()
print(duplicated_daily)
```

Với **daily dataset**, mỗi dòng đại diện cho một cặp *(country, date)*. Việc kiểm tra trùng lặp theo hai cột này trả về giá trị `False`, cho thấy không có dòng nào bị lặp lại trong dữ liệu theo ngày. Điều này xác nhận rằng dữ liệu ghi nhận theo ngày cho mỗi quốc gia là duy nhất và không bị trùng nhập.

```python
duplicated_summary = df_summary.duplicated(subset = ["country"]).any()
print(duplicated_summary)
```

Tương tự, với **summary dataset**, mỗi dòng đại diện cho một quốc gia duy nhất. Kiểm tra trùng lặp theo cột `country` cũng trả về giá trị `False`, đồng nghĩa với việc không có quốc gia nào xuất hiện nhiều lần trong dataset tổng hợp, và tổng số dòng trùng lặp bằng 0.

## 6. The meaning of each column

### 6.1 Daily dataset

```python
df_daily.info()
```

* `date` (*object*): Ngày ghi nhận số liệu.
* `country` (*object*): Tên quốc gia; mỗi dòng tương ứng với một quốc gia tại một ngày cụ thể.
* `cumulative_total_cases` (*float64*): Tổng số ca nhiễm tích lũy đến ngày đó.
* `daily_new_cases` (*float64*): Số ca nhiễm mới trong ngày.
* `active_cases` (*float64*): Số ca đang điều trị.
* `cumulative_total_deaths` (*float64*): Tổng số ca tử vong tích lũy.
* `daily_new_deaths` (*float64*): Số ca tử vong mới trong ngày.

Dữ liệu `daily` có tính chất **chuỗi thời gian**, với mỗi dòng là một quan sát theo ngày của một quốc gia.

### 6.2 Summary dataset


```python
df_summary.info()
```

* `country` (*object*): Tên quốc gia; mỗi dòng là một quốc gia duy nhất.
* `continent` (*object*): Châu lục mà quốc gia thuộc về.
* `total_confirmed` (*int64*): Tổng số ca mắc COVID-19 tích lũy.
* `total_deaths` (*float64*): Tổng số ca tử vong.
* `total_recovered` (*float64*): Tổng số ca hồi phục.
* `active_cases` (*float64*): Số ca đang điều trị hiện tại.
* `serious_or_critical` (*float64*): Số ca nghiêm trọng hoặc nguy kịch.
* `total_cases_per_1m_population` (*int64*): Số ca tích lũy trên 1 triệu dân.
* `total_deaths_per_1m_population` (*float64*): Số ca tử vong trên 1 triệu dân.
* `total_tests` (*float64*): Tổng số xét nghiệm COVID-19 đã thực hiện.
* `total_tests_per_1m_population` (*float64*): Số xét nghiệm trên 1 triệu dân.
* `population` (*int64*): Dân số của quốc gia.

Dataset `summary` có tính chất **tổng hợp**, cung cấp cái nhìn toàn cảnh về mức độ ảnh hưởng của COVID-19 tại từng quốc gia.
