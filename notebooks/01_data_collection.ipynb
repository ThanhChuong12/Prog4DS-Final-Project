### **6.3 Missingness Patterns**

```python
num_rows = df.shape[0]

for col_name in df.columns:
    col = df[col_name]
    missing_count = col.isna().sum()
    if col.dtype == object or col.dtype.name == "category":
        missing_count += col[col.astype(str).str.strip() == ""].shape[0]
    missing_percent = (missing_count / num_rows) * 100
    print(f"{col_name:40}: {missing_percent:.2f}%")
```

Thông qua kiểm tra sơ bộ (`df.info()` và tỷ lệ giá trị thiếu), bộ dữ liệu bộc lộ tình trạng thiếu dữ liệu không đồng đều:

  * **Mức độ nghiêm trọng:** Các biến `Sunshine` ($\approx$48%), `Evaporation` ($\approx$43%), và `Cloud` ($\approx$40%) thiếu gần một nửa số lượng quan sát. Điều này gợi ý sự không đồng nhất về trang thiết bị giữa các trạm (không phải trạm nào cũng đo được nắng/bốc hơi).
  * **Mức độ thấp:** Các biến quan trọng cho dự báo như `Humidity`, `Temp`, `Wind` có tỷ lệ thiếu thấp (\<10%), đảm bảo tính khả thi cho việc xây dựng mô hình.

### **7.2 Potential Research Questions (RQs)**

Bộ dữ liệu được sử dụng để trả lời các câu hỏi nghiên cứu cốt lõi sau:

* **RQ1 (Pressure Dynamics):** Liệu mối quan hệ giữa sụt giảm áp suất khí quyển trong ngày (Pressure9am − Pressure3pm) và xác suất mưa ngày kế tiếp (`RainTomorrow`) có ổn định trên toàn bộ lãnh thổ Úc, hay hiệu lực dự báo của áp suất phụ thuộc phi tuyến vào trạng thái độ ẩm khí quyển và bối cảnh địa lý (Ven biển so với Nội địa)?
* **RQ2 (Seasonality & Indicators):** Sự biến đổi của các đặc trưng khí tượng theo chu kỳ mùa vụ có mối tương quan thế nào với khả năng xảy ra hiện tượng mưa ngày tiếp theo, và đâu là những 'dấu hiệu' nhận biết sớm (early indicators) quan trọng nhất?
* **RQ3 (Thermodynamic Interaction):** Phân tích sự tương tác phi tuyến giữa Biên độ nhiệt ngày (Diurnal Temperature Range - $DTR$) và Thời lượng nắng (Sunshine): Liệu sự kết hợp của hai yếu tố này có tạo ra một trạng thái 'Bất ổn định nhiệt ẩm' giúp dự báo mưa chính xác hơn các chỉ số đơn lẻ, và sự tương tác này thay đổi thế nào qua các nhóm trạm quan trắc khác nhau?
* **RQ4 (Wind Dynamics):** Liệu chỉ số tương tác động lực gió, được xây dựng từ sự kết hợp giữa tốc độ gió giật (`WindGustSpeed`) và độ lệch hướng gió trong ngày (`Diurnal Wind Shift`), có thể đóng vai trò như một chỉ báo dự báo mưa ngày kế tiếp (`RainTomorrow`); hay hiệu lực dự báo của tín hiệu động lực này bị điều biến bởi trạng thái nhiệt – ẩm của khí quyển (đặc trưng bởi độ ẩm và nhiệt độ buổi chiều), cũng như bối cảnh động lực học khu vực (hệ thống gió ven biển so với đối lưu nội địa)?
* **RQ5 (Data Preprocessing):** Phương pháp xử lý missing values nào (Simple Univariate Imputation vs. Multivariate Iterative Imputation) là tối ưu nhất để bảo toàn cấu trúc phân phối thống kê của các đặc trưng có tỷ lệ missing values cao trong dữ liệu (`Sunshine`, `Evaporation`, `Cloud9am`, `Cloud3pm`)?
* **RQ6 (Predictive Modeling):** Trong bối cảnh dữ liệu mất cân bằng (tỷ lệ ngày mưa thấp hơn nhiều so với ngày không mưa), mô hình học máy nào thuộc ba nhóm kiến trúc chính: **Linear** (Logistic Regression), **Bagging** (Random Forest) hay **Boosting** (XGBoost) mang lại hiệu quả cao nhất trong việc tối đa hóa **Recall** (Độ nhạy) để giảm thiểu rủi ro bỏ sót dự báo mưa, trong khi vẫn duy trì chỉ số **F1-Score** ở mức chấp nhận được?
