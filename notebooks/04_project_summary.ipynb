## **3. Limitations**

Mục tiêu: Thể hiện tư duy phản biện (Critical Thinking). Thừa nhận hạn chế để tăng độ tin cậy.

3.1 Hạn chế về Dữ liệu (Dataset Constraints):

Vấn đề điểm trạm (Point-based): Thiếu dữ liệu không gian (Radar/Vệ tinh). Mưa có thể rơi cách trạm 5km nhưng trạm không ghi nhận -> False Negative.

Độ phân giải thời gian (Temporal Resolution): Khoảng cách 6 giờ (9am-3pm) là quá lớn để bắt các cơn dông nhiệt (thường chỉ kéo dài 1-2 giờ).

3.2 Hạn chế về Phương pháp (Methodological Limits):

Thiếu thông tin tầng cao: Chỉ dùng dữ liệu bề mặt (Surface data). Không có thông tin về Dòng xiết (Jet Stream) hay độ ổn định khí quyển (CAPE) – những nguyên nhân gốc rễ của mưa.

3.3 Phạm vi (Scope):

Chưa xét đến tác động của các hiện tượng khí hậu vĩ mô như El Niño / La Niña (ENSO) lên các mối tương quan này.
